Q1) Define the z-statistic and explain its relationship to the standard normal distribution. How is the 
z-statistic used in hypothesis testing

Ans) Z-statistic Definition
The z-statistic (also called the z-score) is a measure that indicates how many standard deviations an element is from the mean of a distribution. It is calculated as:

𝑧 = (𝑋−𝜇)/𝜎


Where:

X = observed value
μ = population mean
σ = population standard deviation

Relationship to the Standard Normal Distribution
The z-statistic transforms data from any normal distribution to a standard normal distribution, which has a mean of 0 and a standard deviation of 1. The standard normal distribution is symmetrical and follows a bell-shaped curve. The z-statistic allows us to compare data points from different distributions on a common scale.

How Z-statistic is Used in Hypothesis Testing
In hypothesis testing, the z-statistic is used to determine whether to reject the null hypothesis. The basic steps are:

State the hypotheses: Define the null hypothesis (H 0) and the alternative hypothesis (H A).

Compute the z-statistic: This compares the observed sample statistic to the population parameter under the null hypothesis.

Find the p-value: The p-value is the probability of observing a test statistic at least as extreme as the one computed, under the null hypothesis. The z-statistic helps determine this by relating the result to a standard normal distribution.

Make a decision:

If the p-value is smaller than the significance level (α), reject the null hypothesis.
Otherwise, fail to reject the null hypothesis.
Example: If the z-statistic is 2, this means the observed value is 2 standard deviations above the mean. Using a z-table or a standard normal distribution, we can calculate the probability (p-value) corresponding to a z of 2 and determine if this is statistically significant based on the predefined significance level.?

Q2)  What is a p-value, and how is it used in hypothesis testing? What does it mean if the p-value is 
very small (e.g., 0.01)

Ans) A p-value (probability value) is a measure used in statistical hypothesis testing to quantify the strength of the evidence against the null hypothesis (H 0). It represents the probability of obtaining test results at least as extreme as the observed results, assuming that the null hypothesis is true.

In formal terms:

The p-value is the probability of observing a test statistic (e.g., z-statistic or t-statistic) at least as extreme as the one calculated from your sample data, under the assumption that the null hypothesis is correct.
How is the p-value used in hypothesis testing?
The p-value is used to decide whether to reject or fail to reject the null hypothesis in hypothesis testing:

Set a significance level (α): Before performing the test, researchers choose a threshold significance level, often 0.05 or 0.01, which defines the level of risk they are willing to take in rejecting the null hypothesis when it is true (Type I error).

Compare the p-value to 𝛼:

If p-value ≤ α, reject the null hypothesis (𝐻 0). This suggests that there is strong evidence against the null hypothesis.
If p-value > 𝛼, fail to reject the null hypothesis. This means there isn’t enough evidence to conclude that the alternative hypothesis (𝐻 𝐴) is true.
What does a very small p-value (e.g., 0.01) mean?
A small p-value indicates that the observed data is very unlikely under the assumption that the null hypothesis is true. Specifically:

p-value = 0.01 means there is a 1% chance that the observed data would occur if the null hypothesis were true. This provides strong evidence to reject the null hypothesis.
Implications:
p-value < 0.05: The results are considered statistically significant, and the null hypothesis is likely false. The smaller the p-value, the stronger the evidence against 𝐻 0.
p-value < 0.01: The results are highly statistically significant, providing very strong evidence against the null hypothesis.
p-value ≈ 0: Extremely small p-values suggest the observed result is highly unlikely under the null hypothesis, further justifying the rejection of 𝐻 0.?

The binomial distribution is used when:

Fixed number of trials (n).
Each trial has two possible outcomes: success or failure.
The probability of success (p) is constant for each trial.
The trials are independent of each other.
It gives the probability of obtaining a certain number of successes in n trials.

Relationship to Bernoulli Distribution:
The Bernoulli distribution is a special case of the binomial distribution with only one trial (𝑛 = 1). Each Bernoulli trial results in either a success (with probability p) or a failure (with probability 1−p).

Ans) Key Properties of the Poisson Distribution:
Discrete Distribution: It describes the probability of a given number of events occurring in a fixed interval (time, space, etc.).
Mean and Variance: Both are equal to 
λ (the average rate of occurrence of the events).

Memoryless: The probability of events in one interval is independent of events in another interval.
Non-negative integers: It models the occurrence of events as non-negative integers (0, 1, 2, ...).

When to Use the Poisson Distribution:
The Poisson distribution is appropriate when:

Events occur randomly and independently in a fixed interval.
The average rate (λ) of event occurrence is known, but the exact timing of events is unknown.
The probability of multiple events happening simultaneously is very small.?

Q6) Define the terms "probability distribution" and "probability density function" (PDF). How does a PDF differ from a probability mass function (PMF)?

Ans) Probability Distribution:
A probability distribution describes how the probabilities of different possible outcomes of a random variable are distributed. It can be either discrete (for countable outcomes) or continuous (for uncountable outcomes over a range).

Probability Density Function (PDF):
A probability density function (PDF) is used for continuous random variables. It describes the likelihood of the random variable taking a specific value. While the probability at any exact point is 0, the PDF can be used to calculate the probability that the variable falls within a range (e.g., between two numbers) by integrating over that range.

The total area under the PDF curve is always 1.
Difference Between PDF and PMF:
PDF: Used for continuous variables. It gives a "density" of probabilities, not direct probabilities.
PMF: Used for discrete variables. It gives the exact probability of each possible outcome (e.g., the probability of rolling a 3 on a die).?

Q7) Explain the Central Limit Theorem (CLT) with example.

Ans) Explanation of the Code:
Population Creation: We generate a large population of 10,000 samples drawn from a uniform distribution between 0 and 100. This distribution is clearly non-normal.

Sampling: We then take 1,000 random samples of size 50 from this population. For each sample, we calculate the mean and store it in the sample_means list.

Visualization: Finally, we plot two histograms:

The first histogram shows the original population distribution, which is uniform.
The second histogram shows the distribution of the sample means, which should approximate a normal distribution due to the CLT.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

population_size = 10000
population = np.random.uniform(low=0, high=100, size=population_size)

sample_size = 50
num_samples = 1000
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(population, bins=30, kde=True, color='blue', stat='density')
plt.title('Population Distribution (Uniform)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=30, kde=True, color='orange', stat='density')
plt.title('Distribution of Sample Means (n=50)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

Q8) Compare z-scores and t-scores. When should you use a z-score, and when should a t-score be applied instead?

Ans) Z-scores vs. T-scores
Z-scores:

Definition: Represents the number of standard deviations a data point is from the mean of a distribution.
Used When:
The population standard deviation (σ) is known.
The sample size is large (n>30).
The underlying distribution is normal.

T-scores:

Definition: Similar to z-scores but used when the sample size is small and the population standard deviation is unknown. It incorporates additional variability due to estimating the population standard deviation from the sample.
Used When:
The population standard deviation is unknown.
The sample size is small (n≤30).
The underlying distribution is approximately normal.
Summary:
Use z-scores for large samples or when the population standard deviation is known.
Use t-scores for small samples or when the population standard deviation is unknown.

Q9) Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample
size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to reject the null hypothesis?

Task: Write Python code to calculate the z-score and p-value for the given data.

Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.

```python
import numpy as np
from scipy import stats

sample_mean = 105
population_mean = 100
population_std_dev = 15
sample_size = 25
significance_level = 0.05
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(z_score))
z_score, p_value
```

Q10) Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python.Generate 1,000 samples and plot the distribution. What is the expected mean and variance?

Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance.

Objective: Understand the properties of a binomial distribution and verify them through simulation.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

n_trials = 10
p_success = 0.6
n_samples = 1000
samples = np.random.binomial(n_trials, p_success, n_samples)
expected_mean = n_trials * p_success
expected_variance = n_trials * p_success * (1 - p_success)
plt.figure(figsize=(10, 6))
sns.histplot(samples, bins=np.arange(-0.5, n_trials + 1.5, 1), kde=False, color='blue', stat='density')
plt.title('Binomial Distribution (n=10, p=0.6)')
plt.xlabel('Number of Successes')
plt.ylabel('Density')
plt.xticks(np.arange(0, n_trials + 1))
plt.grid(axis='y')
plt.show()
expected_mean, expected_variance
```
